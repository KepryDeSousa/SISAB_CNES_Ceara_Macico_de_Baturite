### 1. Scatter + Seaborn 

   "Teste"


```python
# Criando gráfico
trace = go.Scatter(y = AAA['Municipio'],
                   x = AAA['Q1_2020'],
                   mode = 'markers')
trace1 = go.Scatter(y = AAA['Municipio'],
                   x= AAA['Q2_2020'],
                   mode = 'markers')
# Armazenando gráfico em uma lista
data = [trace,trace1]
# Criando Layout
layout = go.Layout(title='Indicadores de Desempenho do Maciço de Baturité',
                   yaxis={'title':'Municipio'},
                   xaxis={'title': 'Quadrimestre'})
# Criando figura que será exibida
fig = go.Figure(data=data, layout=layout, )
# Exibindo figura/gráfico
py.iplot(fig)
```

### 3. Grafico 2020 - 2022 (Valores Totais)

```python
grafico = px.scatter(banco, x= 'Indicador', y=['Total_2020','Total_2021'],# Adicionando a biblioteca com valores para a variavél grafico
                 color='Municipio',                                       #adicionando cores aos valores de municipio
                 hover_data=['Municipio'])                                # adicionando os valores aos pontos flutuantes
grafico.update_layout(title = 'Indicadores de Desempenho do Maciço de Baturité') #configurando titulos dos eixos x e y
grafico.update_yaxes(title = 'Indicadores')
grafico.update_xaxes(title ='Total entre 2020 - 2021 dos indicadores' )
grafico.update_traces(marker = dict(size = 12, line=dict(width = 3)), selector = dict(mode = 'markers')) # responsavél pelo tamanho
grafico.show()
```

### 4. Graficos dos anos 2020-2021 / Quadrimestres de 1 ao 3
